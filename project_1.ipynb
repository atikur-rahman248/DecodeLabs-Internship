```python
import pandas as pd
data = pd.read_csv('/content/Dataset for Data Analytics.csv')
```

```python
data['CouponCode'] = data['CouponCode'].fillna('No Coupon')
```

```python
data
```

missing cuponcode values were replaced with 'no coupon'

```python
data.duplicated().sum()
```

no duplicate rows

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data.info()
```

data type correction: date

```python
text_cols = ['Product','PaymentMethod',
             'OrderStatus','ReferralSource',
             'CouponCode']

for col in text_cols:
    data[col] = data[col].str.strip()
```

```python
data.head()
```

text standardization complete

```python
data['CalculatedTotal'] = data['Quantity'] * data['UnitPrice']
```
