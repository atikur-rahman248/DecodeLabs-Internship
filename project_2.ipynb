distribution of total price

```python
from scipy.stats import skew, kurtosis
def analyze_shape(data):
    s = skew(data)
    k = kurtosis(data)

    plt.figure(figsize=(8, 4))
    sns.histplot(data, kde=True, color='purple')
    plt.title(f"Skewness: {s:.2f} | Kurtosis: {k:.2f}")
    plt.axvline(np.mean(data), color='red', linestyle='--', label='Mean')
    plt.axvline(np.median(data), color='green', linestyle='-', label='Median')
    plt.legend()
    plt.show()

analyze_shape(df['TotalPrice'])
```

here the distribution of total price are right  positive skewed because there is mean > median.






**detect outlier**

IQR Method to identify the outlier of total price

```python
Q1 = np.percentile(df['TotalPrice'], 25)
Q3 = np.percentile(df['TotalPrice'], 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df['TotalPrice'][(df['TotalPrice'] < lower_bound) | (df['TotalPrice'] > upper_bound)]

print(f"IQR: {IQR}")
print(f"Bounds: {lower_bound} to {upper_bound}")
print(f"Identified Outliers: \n{outliers}")

print("Number of Outliers:", len(outliers))
```
